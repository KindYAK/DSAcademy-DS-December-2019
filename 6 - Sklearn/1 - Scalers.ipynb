```python
# Min Max
from sklearn.preprocessing import MinMaxScaler

data = [
    [30, 2, 1000], 
    [45, 6, 1500], 
    [10000, 10, 20000], 
    [78, 5.5, 6000]
]
scaler = MinMaxScaler((0, 100))
scaler.fit(data)
# print(scaler.transform(data))
# print(scaler.transform([[50, 15, 2500]]))
scaler.inverse_transform([[  0.20060181, 162.5,          7.89473684]])
```

```python
# MaxAbs

from sklearn.preprocessing import MaxAbsScaler
X = [[ 1., -1.,  2.],
     [ 2.,  0.,  0.],
     [ 0.,  1., -1.]]
transformer = MaxAbsScaler().fit(X)
print(transformer)
print(transformer.transform(X))
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
data = [[0, 0], [0, 0], [1, 1], [1, 1], [5, 100], [500, 500], [1000000, 10000]]
scaler = StandardScaler()
print(scaler.fit(data))
print(scaler.mean_)
print(scaler.scale_)
print(scaler.transform(data))
print(scaler.transform([[2, 100]]))

scaler.fit_transform(data)
```

```python
import pandas as pd
```

```python
df = pd.read_excel("Schools Data.xlsx")
```

```python
df = df[['total_students_2018', 'diff_teens_2018', 'deficit_places_2018']]
```

```python
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer()
scaler.fit(df)
df = pd.DataFrame(scaler.transform(df), 
                  index=df.index, 
                  columns=df.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
df = df[['location']]
```

```python
encoder = OneHotEncoder()
encoder.fit(df)
df_enc = pd.DataFrame(encoder.transform(df).toarray(), 
                      index=df.index,
                     columns=encoder.categories_)
```
